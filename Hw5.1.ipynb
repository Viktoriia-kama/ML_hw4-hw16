```python
import pandas as pd

url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'

table = pd.read_html(url, match='Коефіцієнт народжуваності в регіонах України', decimal=',', thousands='.')
df = table[0]
df
```

```python
df.head()
```

```python
df.shape
```

```python
import numpy as np
df.replace("—", np.nan, inplace=True)
df
```

```python
df.dtypes
```

```python
for column in df.columns:
    if df[column].dtype == object:
        try:
            df[column] = pd.to_numeric(df[column])
        except ValueError:
            pass
df
df.dtypes
```

```python
a = df.isnull().sum()
sum = len(df)

b = a / sum
b
```

```python
df_without_region = df.drop(columns=['Регіон'])

# Заміна NaN в кожному стовпці на середнє значення цього стовпця
df_without_region.fillna(df_without_region.mean(), inplace=True)
df_without_region
```

```python
# Вилучаємо перший стовпець, який містить назви регіонів
data_without_regions = df.iloc[:, 1:]

# Обчислюємо середні значення для кожного стовпця
mean_values = data_without_regions.mean()

# Замінюємо відсутні дані в кожному стовпці його середнім значенням
df.fillna(mean_values, inplace=True)
df
```

```python
max_2014 = df['2014'].idxmax()
df.loc[max_2014, 'Регіон']
```

```python
import matplotlib.pyplot as plt

born_2019 = df['2019']
regions = df['Регіон']


plt.figure(figsize=(8,5))
plt.bar(regions, born_2019, color='skyblue')
plt.title('Народжуваність по регіонах у 2019 році')
plt.xlabel('Регіон')
plt.ylabel('Народжуваність')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
