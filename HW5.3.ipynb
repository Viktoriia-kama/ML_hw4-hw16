```python
import pandas as pd

df = pd.read_csv('bestsellers with categories.csv')
df
```

```python
df.shape
```

```python
df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
```

```python
df
```

```python
na_counts = df.isna().sum()

print("Кількість пропусків у кожному стовпці:")
print(na_counts)
```

```python
unique_genres = df['genre'].unique()

print("Унікальні значення у колонці 'Genre':")
print(unique_genres)
```

```python
import matplotlib.pyplot as plt
df['price'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')

plt.title('Розподіл цін')
plt.xlabel('Ціна')
plt.ylabel('Кількість')
```

```python
max_price = df['price'].max()

min_price = df['price'].min()

mean_price = df['price'].mean()

median_price = df['price'].median()

print("Максимальна ціна:", max_price)
print("Мінімальна ціна:", min_price)
print("Середня ціна:", mean_price)
print("Медіанна ціна:", median_price)
```

```python
max_rating = df['user_rating'].max()

print("Найвищий рейтинг у датасеті:", max_rating)
```

```python
num_books_with_max_rating = len(df[df['user_rating'] == max_rating])
print("Кількість книг з найвищим рейтингом:", num_books_with_max_rating)
```

```python
max_reviews = df['reviews'].max()

print("Найбільше відгуків у датасеті:", max_reviews)
```

```python
top_50_2015 = df[(df['year'] == 2015)==True]

most_expensive_book = top_50_2015[top_50_2015['price'] == top_50_2015['price'].max()]
most_expensive_book
```

```python
num_books_Fiction_2010 = len(df[(df['year'] == 2010) & (df['genre'] == 'Fiction')])
num_books_Fiction_2010
```

```python
num_books_2010_11 = len(df[((df['year'] == 2010) | (df['year'] == 2011)) & (df['user_rating'] == 4.9)])
num_books_2010_11
```

```python
filtered_books_2015 = df[(df['year'] == 2015) & (df['price'] < 8)]
filtered_books_2015.sort_values(by='price', ascending=True)
```

```python
df.groupby('genre')['price'].agg(['min', 'max'])
```

```python
author_books_count = df.groupby('author')['name'].agg('count').reset_index()

# Перейменування стовпців
author_books_count.columns = ['Автор', 'Кількість книг']
author_books_count
```

```python
author_books_count.shape
```

```python
max = author_books_count['Кількість книг'].idxmax()
author_books_count.loc[max, 'Автор']
```

```python
author_books_count['Кількість книг'].max()
```

```python
author_books_count2 = df.groupby('author')['user_rating'].agg('mean').reset_index()

# Перейменування стовпців
author_books_count2.columns = ['Автор', 'Рейтинг']
author_books_count2
```

```python
min = author_books_count2['Рейтинг'].idxmin()
author_books_count2.loc[min, 'Автор']
```

```python
author_books_count2['Рейтинг'].min()
```

```python
author_books_count
```

```python
# Об'єднання двох датафреймів за стовпцями
merged_df = pd.concat([author_books_count.set_index('Автор'), author_books_count2.set_index('Автор')], axis=1)

# Виведення результату
print("Об'єднаний датафрейм:")
print(merged_df)
merged_df
```

```python
merged_df.sort_values(by=['Кількість книг', 'Рейтинг'], ascending=[True, True])
```
